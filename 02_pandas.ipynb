## Where
WHERE ведет себя как фильтр строк, которые мы хотим вывести

```python
#выбираем строки, которые удовлетворяют условию frame.Birth >= pd.datetime(1985,1,1)
df[df.Birth >= pd.datetime(1985,1,1)]
```

```python
#выбираем строки, удовлетворяющие пересечению условий
df[(df.Birth >= pd.datetime(1985,1,1)) &
   (df.City != 'Москва')]
```

```python
#выбираем строки, удовлетворяющие объединению условий
df[(df.Birth >= pd.datetime(1985,1,1)) |
   (df.City == 'Волгоград')]
```

## Group by
Агрегирования с помощью GROUP BY нужны для конвертации нескольких строк в одну. При этом, во время агрегирования для разных колонок используется разная логика (сумма, среднее, количество)

```python
df.groupby('City').first().reset_index()
```

```python
df.groupby(['City', 'Position']).first().reset_index()
```

```python
df['Money'] = np.random.uniform(low=20000, high=60000, size=(len(df),))
df['Money'] = round(df['Money'], -3)
df['Money'] = df['Money'].astype(int)
df
```

```python
df.groupby('City')['Money'].sum().reset_index()
```

```python
df['Year'] = df['Birth'].dt.year
df.groupby(['City', 'Year']).agg({'Money':np.mean, 'Name': np.ma.count}).reset_index()
```

## Join
Операция JOIN необходима для объединения связанных данных

```python
cities = ['Волгоград', 'Москва', 'Ростов']
population = [1000000, 10000000, 1000000]
city = pd.DataFrame({'City' : cities, 'Population': population})
city
```

```python
pd.merge(df, city, on='City')
```
