## Косинусное растояние
### Счетчик слов

```python
sentences = [
    'привет друг',
    'привет привет',
    'как дела друг',
    'доброе утро',
    'что делаешь'
]
tokens = set(' '.join(sentences).split())
vectors = [[sentence.count(token) for token in tokens]
           for sentence in sentences]
print([(i, word) for i, word in enumerate(tokens)])
vectors
```

```python
for i, sentence in enumerate(sentences):
    print(sentence)
    print(scipy.spatial.distance.cosine(vectors[0], vectors[i]))
    print(scipy.spatial.distance.euclidean(vectors[0], vectors[i]))
    print(scipy.spatial.distance.cityblock(vectors[0], vectors[i]))         
```
